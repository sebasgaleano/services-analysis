```python
# 02_visualization.ipynb

import geopandas as gpd
import folium
from folium.plugins import HeatMap

# Cargar datos preparados
points_gdf = gpd.read_file("../data/asuncion_amenities.geojson")

# Seleccionar dos servicios para superposición
services = ["hospital", "pharmacy"]
colors = {"hospital":"red", "pharmacy":"purple"}

# Crear mapa centrado en Asunción
m = folium.Map(location=[-25.3, -57.6], zoom_start=12)

for service in services:
    filtered = points_gdf[points_gdf['amenity'] == service]
    heat_data = []
    
    for _, row in filtered.iterrows():
        folium.CircleMarker(
            location=[row.geometry.y, row.geometry.x],
            radius=4,
            color=colors[service],
            fill=True,
            fill_opacity=0.7,
            popup=row.get('name', service.title())
        ).add_to(m)
        heat_data.append([row.geometry.y, row.geometry.x])
    
    # Añadir HeatMap por servicio
    if heat_data:
        HeatMap(heat_data).add_to(m)

# Mostrar mapa
m
```
