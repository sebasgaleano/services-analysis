```python
# 01_data_exploration.ipynb

import osmnx as ox
import geopandas as gpd
import os

# Configuración básica
city = "Asuncion, Paraguay"
tags = {"amenity": True}

# Cargar datos
gdf = ox.features.features_from_place(city, tags)

# Explorar columnas y tipos de geometría
print(gdf.columns)
print(gdf.geometry.type.value_counts())

# Filtrar solo puntos para análisis espacial
points_gdf = gdf[gdf.geometry.type == "Point"]

# Revisar tipos de servicios
print(points_gdf['amenity'].value_counts())
# Guardar datos limpios localmente para no recargar OSM cada vez
os.makedirs("../data", exist_ok=True)
points_gdf.to_file("../data/asuncion_amenities.geojson", driver="GeoJSON")
```
