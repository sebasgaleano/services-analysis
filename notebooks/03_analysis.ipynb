```python
# 03_analysis.ipynb

import geopandas as gpd
import numpy as np
from sklearn.cluster import DBSCAN

# Cargar datos
points_gdf = gpd.read_file("../data/asuncion_amenities.geojson")

# Filtrar un servicio para clustering
service = "hospital"
gdf_service = points_gdf[points_gdf['amenity'] == service]

# Convertir coordenadas a array para DBSCAN
coords = np.array(list(zip(gdf_service.geometry.y, gdf_service.geometry.x)))

# Aplicar DBSCAN (eps en radianes si usamos metric='haversine')
db = DBSCAN(eps=0.01, min_samples=3, metric='haversine').fit(np.radians(coords))
gdf_service['cluster'] = db.labels_

# Analizar resultados
print(gdf_service['cluster'].value_counts())

# Guardar resultados con cluster
gdf_service.to_file("../data/hospital_clusters.geojson", driver="GeoJSON")
```
